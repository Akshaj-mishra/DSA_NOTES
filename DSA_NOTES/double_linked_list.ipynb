# **Double linked list**
A **Double linked list** is a two-way list in which all nodes will have two links. This helps in accessing both successor node and predecessor node from the given node position. It provides
bi- directional traversing. Each node contains three fields:
1. Left link.
2. Data.
3. Right link.

The left link points to the predecessor node and the right link points to the successor node. The data field
stores the required data.<br><br>
Many applications require searching forward and backward thru nodes of a list. For example searching
for a name in a telephone directory would need forward and backward scanning thru a region of the whole
list.
